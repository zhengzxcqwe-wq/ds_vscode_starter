# Quick Demo
This notebook shows the expected workflow: data, chart, and import from `src`.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. 生成数据
x = np.linspace(0, 10, 200)       # 从 0 到 10 均匀取 200 个点
y = np.sin(x)                      # 计算 sin(x)

# 2. 简单打印 DataFrame
df = pd.DataFrame({"x": x, "sin(x)": y})
print(df.head())

# 3. 画图
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)')
plt.title("Sine Function")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```
